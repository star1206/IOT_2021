##[실습1]데이터 가져오고 저장하기

- KBO 홈페이지 기록-기록실 페이지에 접속
  https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx
- 선수기록-타자 선택 후 주소창에 있는 주소 복사

```
# 타자기록 크롤링해서 가져오기
import pandas as pd
tables=pd.read_html('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx')
tables[0].to_excel('KBO타자기록.xlsx')     #웹크롤링한 데이터 파일을 엑셀파일로 저장한 후 제출
tables[0]
```

```
df=pd.DataFrame(tables[0],columns=['순위','선수명','팀명','AVG'])  #필요한 컬럼만 데이터프레임에 저장
df
```

2. 데이터분석 하기

##[실습2]3할타자 명단 출력하기

```
#3할 타자 현황
df[df['AVG']>0.3]
```

##[실습3]팀별 3할 타자 보유현황 그래프 그리기

```
#팀별 3할 타율 타자 보유 현황(그래프)
df[df['AVG']>0.3]['팀명'].value_counts().sort_index().plot.bar()
```
