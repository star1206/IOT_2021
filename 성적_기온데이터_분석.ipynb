````python
# 연도와 평균기온을 저장
import csv
f = open('/content/drive/MyDrive/data/30902김동현 - 2000_2022_temp.csv','r')
lines = csv.reader(f)
header = next(lines)
year = []
temp = []
for line in lines:
  if '11-17' in line[0]:
    year.append(line[0][:4])
    temp.append(float(line[2]))
    #print(line)
  #print(line)
print(year)
print(temp)
f.close()
````

### 3)데이터 시각화하기1

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,5))
plt.plot(year, temp)
plt.title('매해 수능일의 기온 변화')
plt.xlabel('년도')
plt.ylabel('평균기온')
plt.show()
````

## 수능일의 기온 변화를 그래프로 그리기2(최고기온,최저기온 포함하기)

### 1) 최고기온 최저기온 데이터 추출

````python
import csv
f = open('/content/drive/MyDrive/data/30902김동현 - 2000_2022_temp.csv','r')
lines = csv.reader(f)
header = next(lines)
year = []
temp = []
high = []
low = []
for line in lines:
  if '11-17' in line[0]:
    year.append(line[0][:4])
    temp.append(float(line[2]))
    high.append(float(line[4]))
    low.append(float(line[3]))
    #print(line)
  #print(line)
print(year)
print(temp)
print(high)
print(low)
f.close()
````

### 2) 데이터 시각화하기

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,5))
plt.plot(year, temp, label='평균기온')
plt.plot(year,high, label='최고기온')
plt.plot(year,low, label='최저기온')
plt.title('매해 수능일의 기온 변화')
plt.xlabel('년도')
plt.ylabel('기온')
plt.legend()
plt.show()
````
